```python
sent_classifier = APC.SentimentClassifier(r"C:\Users\MY LAPTOP\OneDrive\Documents\GitHub\ABSA_Finance\Models\LSA_T", auto_device=True)
```

```python
import time
```

```python
examples = [
    "The [B-ASP]food[E-ASP] was good, but the [B-ASP]service[E-ASP] was terrible. $LABEL$ Positive, Negative",  # if you pass $LABEL$ in the end of the sentence, it will only be used for prediction checking
    # "The [B-ASP]food[E-ASP] was terrible, but the [B-ASP]service[E-ASP] was good. $LABEL$ Negative, Positive",
    # "The [B-ASP]food[E-ASP] was so-so, and the [B-ASP]service[E-ASP] was terrible. $LABEL$ Neutral, Negative",
]

start_time = time.time()
for ex in examples:
    sent_classifier.predict(
        text=ex,
        print_result=True,
        ignore_error=True,  # ignore an invalid example, if it is False, invalid examples will raise Exceptions
        eval_batch_size=32,
    )
end_time = time.time()
print(f"Single predictions completed in {end_time - start_time} seconds.")

```
