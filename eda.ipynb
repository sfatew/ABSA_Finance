```python
import pandas as pd
import json
import ast

# Assuming 'df' is your DataFrame
# df = pd.read_csv("your_file.csv") # Uncomment and use your file path if needed

def get_sentiment_counts(df, column_name='Decisions'):
    all_sentiments = []
    
    for index, row in df.iterrows():
        val_str = row[column_name]
        if pd.isna(val_str):
            continue
            
        try:
            # Try parsing as JSON first
            decision_dict = json.loads(val_str)
        except (json.JSONDecodeError, TypeError):
            try:
                # Fallback to literal_eval for Python dictionary strings
                decision_dict = ast.literal_eval(val_str)
            except:
                continue # Skip unparseable rows
        
        # Collect the values (sentiments) from the dictionary
        if isinstance(decision_dict, dict):
            all_sentiments.extend(decision_dict.values())

    # Count the frequencies
    return pd.Series(all_sentiments).value_counts()

# Execute the counting
sentiment_stats = get_sentiment_counts(data)

print("Sentiment Statistics:")
print(sentiment_stats)
```

```python
# Create a new column 'title_length' containing the length of each string
data['title_length'] = data['Title'].str.len()

# Sort by length in descending order and select the top 10
# We display both the title and its length
top_10_longest = data.nlargest(10, 'title_length')[['Title', 'title_length']]

print("Top 10 Longest Titles:")
print(top_10_longest)
```
